```python
## cleaning harvested area:
harvested['Value'] = harvested['Value'].str.replace(',','').astype(float)
harvested = harvested.loc[~harvested['Value'].isnull()]
harvested = harvested.loc[~harvested['County ANSI'].isnull()]
harvested['GEOID'] = harvested['State ANSI'].astype(int).astype(str).str.zfill(2) +harvested['County ANSI'].astype(int).astype(str).str.zfill(3)
harvested['harvested_hac'] = harvested['Value'] * 1/0.404686 #changing acre -> hac but does not matter actually
harvested = harvested.rename(columns={'Year':'year'})
print(harvested.groupby(['GEOID'])['year'].nunique().unique())
harvested.head()
```

```python
def predict(df, fit_model, case , name, crop_yld_name, opt_bin):
    new_df = df.copy()
    if crop_yld_name == 'spr_wheat_yld':
        include_col = ['GEOID', 'year', 'bin1v1', 'bin1v2', 'bin1v3', 'bin1v4', 'bin1v5',
       'bin1v6', 'bin2v1', 'bin2v2', 'bin2v3', 'bin2v4', 'bin2v5', 'bin2v6',
       'bin3v1', 'bin3v2', 'bin3v3', 'bin3v4', 'bin3v5', 'bin3v6', 'bin4v1',
       'bin4v2', 'bin4v3', 'bin4v4', 'bin4v5', 'bin4v6', 'bin5v1', 'bin5v2',
       'bin5v3', 'bin5v4', 'bin5v5', 'bin5v6', 'bin6v1', 'bin6v2', 'bin6v3',
       'bin6v4', 'bin6v5', 'bin6v6']
    else:
        include_col = ['GEOID', 'year', 'bin1v1', 'bin1v2', 'bin1v3', 'bin1v4', 'bin1v5',
       'bin1v6', 'bin2v1', 'bin2v2', 'bin2v3', 'bin2v4', 'bin2v5', 'bin2v6',
       'bin3v1', 'bin3v2', 'bin3v3', 'bin3v4', 'bin3v5', 'bin3v6', 'bin4v1',
       'bin4v2', 'bin4v3', 'bin4v4', 'bin4v5', 'bin4v6', 'bin5v1', 'bin5v2',
       'bin5v3', 'bin5v4', 'bin5v5', 'bin5v6', 'bin6v1', 'bin6v2', 'bin6v3',
       'bin6v4', 'bin6v5', 'bin6v6', 'percent_n_b_temp']

    #### making no exposure to hydroclimatic hazard ############################################################################
    new_df[case]=0
    #### predicting yields without hydroclimatic hazard ########################################################################
    df[name+'_pred_yld'] = fit_model.predict(new_df[include_col])
    #### finding gap between yields without specific hazard and "pred" (yields with all hydroclimatic hazards) #################
    #### name_pred_yld is the yield when that specific hazard was removed ######################################################
    df[name+'_gap'] =  df[name+'_pred_yld'] - df['pred']
    #### Cases where no exposure to hydroclimatic hazards happen, zero or negative outputs close to zero. ###################### 
    #### We do not remove thoese counties since pred has all county info and use them as our base ##############################
    df.loc[df[name+'_gap'] < 0, name+'_gap'] = 0.0
    return df
```

```python
for idx, key in enumerate(hazards.keys()):
    hypo_df = predict(hypo_df, fit_both_full_1, hazards[key], key, crop_yld_name, opt_bin)
hypo_df = hypo_df.drop(columns=hypo_df.columns[hypo_df.columns.str.contains('bin')])
```

```python
hypo_df['crop_name'] = crop_name
hypo_df.to_csv('../3_outputs/'+crop_yld_name+'_geoid_level_yld_gap.csv')
```

```python
### nationally averaged yield gaps weighted by harvested areas of all counties for each crop:
hypo_df['national_harvested'] = hypo_df.groupby(['year'])['harvested_hac'].transform(sum)
hypo_df['national_weight'] = hypo_df['harvested_hac'] / hypo_df['national_harvested']
year_df = pd.DataFrame()
wyld_list =list(hypo_df.columns[hypo_df.columns.str.contains('gap')]) + ['kgha_'+crop_yld_name, 'pred']
for idx, key in enumerate(wyld_list):
    hypo_df[key+'_wyld'] = hypo_df[key] * hypo_df['national_weight']
    if idx == 0:
        year_df = hypo_df.groupby(['year'])[key+'_wyld'].sum().reset_index()
    else:
        year_df = pd.merge(year_df, hypo_df.groupby(['year'])[key+'_wyld'].sum().reset_index(), 
                           on=['year'], how='outer')
## convert kg/ha to tonne/ha:
year_df.loc[:, year_df.columns.str.contains('wyld')] = year_df.loc[:, year_df.columns.str.contains('wyld')] * 1e-3
```

```python
### calculating the yield loss in percent which incdicates how much each hydroclimatic hazard induced yield loss out of the counterfauctual scenario ###
for idx, key in enumerate(year_df.columns[year_df.columns.str.contains('gap_wyld')]):
    year_df[key+'_%'] = (year_df[key] / (year_df['all_gap_wyld'] + year_df['pred_wyld'])) * 100 * -1 
year_df
```

```python
year_df.to_csv('../3_outputs/'+crop_name+'_weighted_average_yield_loss_at_annual_national_level.csv')
```
